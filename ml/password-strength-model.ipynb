```python
import joblib
import xgboost as xgb
from sklearn.feature_extraction.text import TfidfVectorizer

model_clone = joblib.load('password-strength-export.pkl')
```

```python
def word(password):
    character=[]
    for i in password:
        character.append(i)
    return character

tdif = TfidfVectorizer(tokenizer=word)

tdif.fit(["2241"])

data = tdif.transform(["2241"])
```

```python
output = model_clone.predict(data)
```
